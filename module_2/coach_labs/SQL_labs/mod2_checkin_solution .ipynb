2. How many authors are called David?  

```python
c.execute("""SELECT count(*) as num_rows FROM author
WHERE author LIKE 'David%'; """) 
```

4. Use both the author table and books table to 
say: how many copies were sold by authors named David 


    a. how many copies for each of the Davids
    
    
    b. how many copies in total for all Davids

### A

```python
c.execute("""SELECT author, sum(num_copies) as num_copies  FROM author
LEFT JOIN books ON books.books = author.book
WHERE author LIKE 'David%'
GROUP BY author; """) 
```

```python
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

### B

```python
c.execute("""SELECT sum(num_copies) as total_books_sold_by_Davids FROM 

(SELECT author, sum(COALESCE(num_copies, 0)) as num_copies  FROM author
LEFT JOIN books ON books.books = author.book
WHERE author LIKE 'David%'
GROUP BY author)



; """) 
```
