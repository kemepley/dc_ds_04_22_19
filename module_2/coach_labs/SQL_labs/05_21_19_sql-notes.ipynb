```python
# How many rows are in the author and books tables?

c.execute('''SELECT COUNT(author)
             From author''').fetchall()

c.execute('''SELECT COUNT(books)
             From books''').fetchall()
```

```python
# How many authors (share) didn't sell some of their book?
c.execute('''SELECT *
             FROM books
             WHERE num_copies ="None"''').fetchall()
```

```python
# How many duplicates are there in the author table?
# c.execute('''SELECT COUNT(*)
#              FROM (SELECT author, COUNT(book) as count_books
#                    FROM author
#                    GROUP BY author, book
#                    HAVING count_books > 1)''').fetchall()

pd.read_sql_query('''SELECT COUNT(DISTINCT(author)) as distinct_auth, COUNT(author)
             FROM author''', conn)
```

```python
# How many authors (share) didn't sell any book?
c.execute('''SELECT SUM(num_copies)
             FROM author
             LEFT JOIN books
             ON author.book = books.books
             GROUP BY author
             HAVING num_copies = 0''').fetchall()
```

```python
pd.read_sql_query('''SELECT *
             FROM author
             LEFT JOIN books
             ON author.book = books.books
             ORDER BY author
             ''', conn)
```

```python
# How many books in total each author has sold?
c.execute('''SELECT author.author, (SUM)books.num_copies
             FROM author
             LEFT JOIN books
             ON author.book = books.books
             GROUP BY author''').fetchall()
```
