### Use SQL to answer the following questions: 

1.  How many rows are in the 'author' database? 

2. How many authors are called David?  

3. How many books each author wrote? 

4. Use both the author table and books table to 
say: how many copies were sold by authors named David 


    a. how many books for each of the Davids
    
    
    b. how many books in total for all Davids

```python
# How many rows are in the 'author' database?
pd.read_sql_query('''SELECT COUNT(*) as Count_of_Rows
                     FROM author''', conn)
```

```python
# How many authors are called David?
pd.read_sql_query('''SELECT COUNT(*) as Count_of_Davids
                     FROM author 
                     WHERE author LIKE "David%"''', conn)
```

```python
# How many books each author wrote?
pd.read_sql_query('''SELECT a.author, COUNT(b.books) as Total_Books_Sold 
                     FROM author AS a
                     JOIN books AS b
                     ON a.book = b.books
                     GROUP BY a.author
                     ORDER BY COUNT(b.books) DESC''', conn)
```

```python
# how many books for each of the Davids
pd.read_sql_query('''SELECT a.author as Author_David, SUM(b.num_copies) as Total_Books_Sold
                     FROM author AS a
                     JOIN books AS b
                     ON a.book = b.books
                     WHERE a.author LIKE "David%"
                     GROUP BY a.author
                     ORDER BY COUNT(b.num_copies) DESC''', conn)

                    
```

```python
# how many books in total for all Davids
pd.read_sql_query('''SELECT SUM(b.num_copies) as Total_Books_Sold_By_Davids
                     FROM author AS a
                     JOIN books AS b
                     ON a.book = b.books
                     WHERE a.author LIKE "David%"''', conn)
```
