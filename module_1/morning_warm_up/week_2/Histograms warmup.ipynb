```python
### Import packages 
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
```

```python
### 1. Create fake data from normal distribution with 100 observations
import numpy as np
fakedata = np.random.normal(size=100)
fakedata
```

```python
### 2. Plot a histogram to see the distribution using matplotlib

plt.hist(fakedata)
```

```python
### 3.switch to seaborn and add a smooth distribution line on top of histogram 

sns.distplot(fakedata)
```

```python
### 4. add titles to plots and save them in your visualizations folder using python commands

ax = plt.hist(fakedata)
plt.title("Fake Title for Fake Data, Matplotlib")

plt.savefig("fake_data_1.png")
```

```python
sns.distplot(fakedata).set_title("Fake Title for Fake Data, Seaborn")
plt.savefig("fake_data_2.png")
```
