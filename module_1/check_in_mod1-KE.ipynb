4. Write a function to process the recipe dictionary. This function should loop through the recipes and return a *list* of ingredients for each recipe.

```python
myDict = {}
myDict['lemon bars'] = {'lemons': '6', 'sugar': '3 cups'}
myDict['key lime pie'] = {'limes': '8', 'sugar': '2 cups'}
myDict['apple pie'] = {'apples': '7', 'sugar': '1 cup'}
myDict
```

```python
ingredients = []
for i in myDict:
    for j in myDict[i].keys():
        ingredients.append(j)
ingredients
```

### Linear regression 

1. What are the assumptions of linear regression? 

Linearity
Normality (of the residuals)
Heteroscedacity

2. Explain the reasoning for each assumption

If the data isn't linear, you might need a different model. If the distribution of the residuals is skewed, then a linear model might be a good predictor for some values but not others (the ones that aren't evenly distributed).

3. Explain the concept of correlation. What is the covariance? Why don't we use covariance? 

Two variables covary if a change in one correlates with a change in the other. 

4. Should we use the correlation to decide on variables to delete from our model?

Multiple regression assumes that we can show how much a change in one feature affects the target holding the other features constant. If features affect each other, then we can't make that assumption.

5. What is R squared? How can it help us in comparing among models? 
What is the R squared of the baseline model?

R2 tells us how much variance in the target is explained by a change in x. It compares the errors of the regression line to the baseline (a line that always predicts the mean of y for any x). 

6. Explain cross-validation. Why do we use it?   
[stretch] Create a function that does cross validation? (use sklearn, numpy etc.)

Cross validation helps us prevent misleading results due to performing only a single train/test split. If you only do one split, you might get a random sample that isn't representative (E.g. it has all of the lowest values in the distribution and none of the highest). If you train your data on that and only that, you won't get an accurate model.

7. How can we evaluate our model besides R squared? and why would we want to? (hint: loss function)

Mean absolute error or mean squared error measure error while preventing positive and negative errors from cancelling each other out.
