### Import packages 

```python
from __future__ import division
import numpy as np
import pandas as pd
```

### Create chompapia data

```python
ages = np.round(np.random.uniform(20, 85, 1000), 2)
```

```python
min_systol_blood_pressure = np.round(np.random.uniform(75, 150, 1000))
```

```python
heart_rate =  np.round(np.random.uniform(15, 160, 1000))
```

```python
data = pd.DataFrame({'age':ages, 'systolic_blood':min_systol_blood_pressure, 'heart_rate': heart_rate})
```

```python
data
```

### Our tree

```python
for patient_num in range(data.shape[0]):
    if (data.loc[patient_num, 'systolic_blood'] > 120):
        data.loc[patient_num, 'label'] = 'chompapia'
        continue
    elif (data.loc[patient_num, 'age'] > 62):
        data.loc[patient_num, 'label'] = 'no chompapia'
        continue
    elif ((data.loc[patient_num,'heart_rate'] > 135) | (data.loc[patient_num,'heart_rate'] < 35)):
        data.loc[patient_num, 'label'] = 'chompapia'
        continue
    else:
        data.loc[patient_num, 'label'] = 'no chompapia'
```

### Create true labels

```python
for patient_num in range(data.shape[0]):
    p = np.random.uniform(0, 1)
    if ((p <= 0.7) & (data.loc[patient_num, 'systolic_blood'] > 120)):
        data.loc[patient_num, 'tr_label'] = 'chompapia'
        continue
    elif ((p <= 0.87) & (data.loc[patient_num, 'age'] > 62)):
        data.loc[patient_num, 'tr_label'] = 'no chompapia'
        continue
    elif ( (p <= 0.72) & ((data.loc[patient_num,'heart_rate'] > 135) | (data.loc[patient_num,'heart_rate'] < 35))):
        data.loc[patient_num, 'tr_label'] = 'chompapia'
        continue
    else:
        data.loc[patient_num, 'tr_label'] = 'no chompapia'
```

### Calculate accuracy/FPR/FNR

```python
accuracy = sum(data['label'] == data['tr_label']) / len(data['label'])
```

```python
accuracy
```

```python
fp = sum((data['label'] == 'chompapia') & (data['tr_label'] != 'chompapia'))
```

```python
tn = sum((data['label'] == 'no chompapia') & (data['tr_label'] == 'no chompapia'))
```

```python
fpr = fp / (fp + tn)
```

```python
fpr
```

```python
fn = sum((data['label'] != 'chompapia') & (data['tr_label'] == 'chompapia'))
```

```python
tp = sum((data['label'] == 'chompapia') & (data['tr_label'] == 'chompapia'))
```

```python
fnr = fn / (fn + tp)
```
